```python
customers=pd.read_csv(r"C:\Users\tarak\OneDrive\Documents\olist_customers.csv")
order_items=pd.read_csv(r"C:\Users\tarak\OneDrive\Documents\olist_order_items.csv")
orders=pd.read_csv(r"C:\Users\tarak\OneDrive\Documents\olist_orders.csv")
sellers=pd.read_csv(r"C:\Users\tarak\OneDrive\Documents\olist_sellers.csv")
order_reviews=pd.read_csv(r"C:\Users\tarak\OneDrive\Documents\olist_order_reviews.csv")
products=pd.read_csv(r"C:\Users\tarak\OneDrive\Documents\olist_products.csv")
payments=pd.read_csv(r"C:\Users\tarak\OneDrive\Documents\olist_order_payments.csv")
```

```python
tables={
    "Orders":orders,
"Customers":customers,
"Products":products,
"Sellers":sellers,
"Order_Items":order_items,
"Order_Reviews":order_reviews,
"Payments":payments}
for name, df in tables.items():
    print(f"\n--- {name} ---\n")
    print("\nInfo:")
    print(df.info())
    print("\nMissing values:")
    print(df.isnull().sum())



    
```

```python
order_reviews['review_comment_message']=order_reviews['review_comment_message'].fillna("No Comment")
order_reviews['review_comment_title'] = order_reviews['review_comment_title'].fillna("No Comment Title")
```

```python
products['product_category_name']=products['product_category_name'].fillna("Unknown")
```

```python
products['product_name_lenght'] = products['product_name_lenght'].fillna(0)
products['product_description_lenght'] = products['product_description_lenght'].fillna(0)
products['product_photos_qty'] = products['product_photos_qty'].fillna(0)
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'],dayfirst=True)
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'],dayfirst=True)
orders['order_approved_at'] = orders['order_approved_at'].fillna(orders['order_purchase_timestamp'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'], dayfirst=True)
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'], dayfirst=True)
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'], dayfirst=True)
```

```python
orders_clean = orders[orders['order_delivered_customer_date'].notnull()].copy()
```

```python
orders_clean['order_delivered_customer_date'].isnull().sum()
```

```python
df=orders_clean.merge(order_items,on="order_id",how="inner")
```

```python
df=df.merge(products,on="product_id",how="left")
df = df.merge(customers, on="customer_id", how="left")
df = df.merge(order_reviews[['order_id','review_score']], on="order_id", how="left")
```

```python
df = df.merge(payments, on="order_id", how="left")
```

```python
df['revenue']=df['price']+df['freight_value']
```

```python
df['delivery_days'] = (
    df['order_delivered_customer_date'] - df['order_purchase_timestamp']
).dt.days
```

```python
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')
```

```python
monthly_orders = df.groupby('order_month').size()
plt.figure(figsize=(8,4))
monthly_orders.plot(kind='line', marker='o')
plt.title("Monthly Order Trend")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.grid()
plt.tight_layout()
plt.show()
```

1️⃣ Monthly Order Trend

Insight

The monthly order trend shows how customer demand changes over time. Variations across months indicate seasonal buying behavior and possible business growth or decline periods.

```python
monthly_revenue=df.groupby('order_month')['revenue'].sum()
monthly_revenue.plot(figsize=(8,2))
plt.title("Monthly Revenue Trend")
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

2️⃣ Monthly Revenue Trend

Insight

Revenue trends generally follow the order volume pattern, indicating that sales performance is primarily driven by the number of orders rather than significant changes in order value.

```python
customers_df=pd.merge(orders_clean,customers,on="customer_id",how="left")
state_orders=customers_df['customer_state'].value_counts()
state_orders.head(10).plot(kind="bar",figsize=(8,4))
plt.title("Top 10 States by Orders")
plt.ylabel('Number of Orders')
plt.show()
plt.tight_layout()
```

3️⃣ Top States by Orders

Insight

A small number of states contribute a large portion of total orders, showing that customer demand is geographically concentrated. These regions represent key markets for business operations and expansion.

```python
repeat=df.groupby('customer_id')['order_id'].nunique()
print("Repeat customers:", (repeat > 1).sum())
```

4️⃣ Repeat Customers

Insight

The presence of repeat customers indicates a certain level of customer retention. However, comparing repeat vs total customers helps understand whether the business relies more on new customers or loyal ones.

```python
df.groupby('product_category_name')['revenue'].sum().sort_values(ascending=False).head().plot(kind="bar",figsize=(8,4))
plt.title('Top Categories by Revenue')
plt.ylabel('Revenue per category')
plt.xticks(rotation=45)
plt.show()
```

5️⃣ Top Categories by Revenue

Insight

Revenue is unevenly distributed across product categories, with a few categories contributing the majority of total sales. This helps identify high-performing segments for inventory and marketing focus.

```python
top_sellers=df.groupby('seller_id')['revenue'].sum().sort_values().head(10)
top_sellers.index = [f"Seller {i+1}" for i in range(len(top_sellers))]
top_sellers.plot(kind='bar', figsize=(10,5))
plt.title('Top Sellers by Revenue')
plt.xlabel("Seller ID")
plt.ylabel("Revenue")
plt.show()
plt.tight_layout()
```

6️⃣ Top Sellers by Revenue

Insight

Revenue is concentrated among a small number of sellers, indicating dependency on top-performing sellers. This concentration may create operational risk if key sellers underperform or leave the platform.

```python
df['delivery_days'].hist(bins=30)
plt.title("Delivery Time Distribution")
plt.title("Distribution of Delivery Time (in Days)")
plt.xlabel("Delivery Days")
plt.ylabel("Frequency")
plt.show()
```

7️⃣ Delivery Time Distribution

Insight

Most orders are delivered within a limited time range, but the presence of longer delivery times indicates operational variability that may impact customer experience.

```python
df.groupby('delivery_days')['review_score'].mean().head(20).plot()
plt.title("Delivery Time vs Review Score")
plt.xlabel("Delivery Days")
plt.ylabel("Average Review Score")
plt.show()
```

8️⃣ Delivery Time vs Review Score

Insight

Customer satisfaction tends to decrease as delivery time increases, suggesting that faster deliveries are an important factor in maintaining higher review scores.

```python
df['payment_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Methods")
plt.show()
```

9️⃣ Payment Method Distribution

Insight

Credit cards dominate the payment methods, indicating strong customer preference for card-based transactions and installment-based purchasing behavior.

```python
aov = df.groupby('order_id')['revenue'].sum().mean()
print("Average Order Value:", aov)
```

🔟 Average Order Value (AOV)

Insight

The Average Order Value represents the typical amount spent per order and helps measure customer purchasing power and overall transaction size.

```python
df.groupby('order_id')['revenue'].sum().hist(bins=50)
plt.title("Order Value Distribution")
plt.xlabel("Order Value (Total Amount)")
plt.ylabel("Number of Orders")
plt.show()
```

1️⃣1️⃣ Order Value Distribution

Insight

Most orders fall within the lower value range, while a small number of high-value orders contribute significantly to total revenue. This indicates a right-skewed spending distribution.

```python

customer_value = df.groupby('customer_id')['revenue'].sum()
top_customers = customer_value.sort_values(ascending=False)
top_10_percent = int(len(top_customers) * 0.1)

top_share = top_customers.head(top_10_percent).sum() / top_customers.sum()
print("Top 10% customers contribute:", top_share)
```

1️⃣2️⃣ Customer Revenue Concentration

Insight

A small percentage of customers contributes a disproportionately large share of total revenue, highlighting the importance of high-value customers to business performance

```python
df['is_late'] = df['order_delivered_customer_date'] > df['order_estimated_delivery_date']
late_rate = df['is_late'].mean()
print("Late delivery rate:", late_rate)
```

1️⃣3️⃣ Late Delivery Rate

Insight

The late delivery rate measures operational reliability. A higher percentage indicates potential issues in logistics and fulfillment that may affect customer satisfaction.

```python
df['day_of_week'] = df['order_purchase_timestamp'].dt.day_name()
df['day_of_week'].value_counts().plot(kind='bar')
plt.title("Orders by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

1️⃣4️⃣ Orders by Day of Week

Insight

Order volume varies across the week, revealing customer purchasing patterns. Identifying peak days helps in planning operations, inventory, and marketing activities.
